```python
import tinker
from tinker.types import SamplingParams
import os
from dotenv import load_dotenv
from tinker_cookbook import tokenizer_utils
from tinker import types
```

```python
load_dotenv()
```

```python
service_client = tinker.ServiceClient()
print("Available models:")
for item in service_client.get_server_capabilities().supported_models:
    print("- " + item.model_name)
```

```python
MODEL_NAME = 'Qwen/Qwen3-8B'
```

```python
service_client = tinker.ServiceClient()
```

```python
sampling_client = service_client.create_sampling_client(base_model=MODEL_NAME)
```

```python
tokenizer = tokenizer_utils.get_tokenizer(MODEL_NAME)
```

```python
text_prompt = "Explain the concept of quantum entanglement in simple terms."
prompt_tokens = tokenizer.encode(text_prompt)
model_input = types.ModelInput.from_ints(prompt_tokens)
```

```python
model_input
```

```python
sampling_params = types.SamplingParams(
    max_tokens=200,
    temperature=0.7,
    top_p=0.95,
    stop=["<|im_end|>"]
)
```

```python
future = sampling_client.sample(
    prompt=model_input, 
    sampling_params=sampling_params, 
    num_samples=1
)
```

```python
response = future.result()
```

```python
for i, sequence in enumerate(response.sequences):
    decoded_text = tokenizer.decode(sequence.tokens)
    print(f"Sample {i+1}: {decoded_text}")
```
